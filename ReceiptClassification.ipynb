```python
# Define the list 'con' with sample data
con = ["Receipt", "Price", "Item", "Total", "Credit"]

# List of keywords to check
keywords = ["TAX", "price", "CHANGE", "Balance", "TOTAL", "credit"]

# Variable to count occurrences
c = 0

# Loop through each keyword
for keyword in keywords:
    # Convert the keyword to lowercase for case-insensitive comparison
    keyword_lower = keyword.lower()
    # Check if any element in 'con' matches the lowercase keyword
    if any(keyword_lower == element.lower() for element in con):
        print("It is a receipt")
        c += 1
        break

# If no keywords are found
if c == 0:
    print("It's not a receipt")
```
